### 1.2. Funções para análise dos dados

```python
def aquisicao_dados(caminho_arq: Path,
                    mapeamento_dados: dict) -> pd.DataFrame:
    """
    Dado um caminho para um arquivo csv, lê o data set e retorna um
    data frame.
    
    Argumentos
    ----------
    caminho_arq: o caminho de um arquivo csv
    
    mapeamento_dados: um dicionário que contém um mapeamento dos dados
                      das colunas de cada arquivo csv

    Return
    ------
    Um data frame com os dados do aquivo csv
    """

    ## cria o nome do arquivo sem a extenção csv, pois ele será
    ## usado como chave para encontrar o mapeamento dos dados
    arquivo_sem_extencao = caminho_arq.name.split('.')[0]
    dataset = pd.read_csv(caminho_arq,
                          dtype=mapeamento_dados[arquivo_sem_extencao],
                          date_format="%Y-%m-%d %H:%M:%S")

    ## cria uma lista contendo apenas as colunas que possuam datas ou
    ## horários a fim de convertê-las adequadamente
    colunas_data_horario = [col for col in dataset.columns
                            if "date" in col or "time" in col
                            or "order_approved_at" in col]

    if len(colunas_data_horario) != 0:
        for col in colunas_data_horario:
            dataset[col] = pd.to_datetime(dataset[col],
                                          format="%Y-%m-%d %H:%M:%S")

    return dataset
```

```python
def graficos_vars_numericas(dados: pd.DataFrame,
                            colunas: list = []) -> None:
    """
    Cria histogramas e boxplots para as variáveis numéricas de um
    data frame.

    Argumentos
    ----------
    dados: um data frame contendo os dados a serem plotados.

    colunas: uma lista contendo quais colunas devem ser usados. É um
             argumento opcional, se não for passsado, então são
             utilizadas todas as colunas numéricas do data frame
    """

    if colunas == []:
        colunas = dados.select_dtypes(include='number').columns.to_list()

    print(f'Serão plotados histogramas e boxplots para as colunas:'
          f'\n{colunas}')

    ## o número de plots é len(colunas) * 2, pois iremos plotar um
    ## histograma e um boxplot para cada variável
    numero_plots = len(colunas) * 2

    fig, ax = plt.subplots(numero_plots // 2,
                           2,
                           figsize=(10, 14))

    ## é executado o ravel para deixar mais conveniente de iterar
    ## os elementos dessa estrutura
    ax = ax.ravel()
    i = 0

    for coluna in colunas:
        sns.histplot(data=dados, x=coluna, ax=ax[i])
        ax[i].set_title(f'Histograma - {coluna}')
        ax[i].set_xlabel(f'Valores - {coluna}')
        ax[i].set_ylabel(f'Frequência')

        sns.boxplot(data=dados, x=coluna, ax=ax[i+1])
        ax[i+1].set_title(f'Boxplot - {coluna}')
        i += 2

    plt.tight_layout()
```

## 2.2 Qual a distribuição do tempo de entrega dos pedidos?

```python
mascara = (ordens['order_status'] == 'delivered') & (ordens['order_delivered_customer_date'] >= ordens['order_approved_at'])
ordens_entregues = ordens[mascara]
tempo_entrega = (ordens_entregues['order_delivered_customer_date'] - ordens_entregues['order_approved_at']).dt.days
tempo_entrega
```

```python
tempo_entrega.describe().T
```

```python
## para analisar a assimetria da distribuição
tempo_entrega.skew()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12, 6))

sns.histplot(data=tempo_entrega,
             ax=ax[0])
ax[0].set_title('Histograma do tempo de entrega')
ax[0].set_xlabel('Tempo de entrega')
ax[0].set_ylabel('Frequência')

sns.boxplot(data=tempo_entrega,
             ax=ax[1])
ax[1].set_title('Boxplot do tempo de entrega')

plt.tight_layout()
plt.show()
```

Antes de entrarmos na análise, gostaríamos de dizer algo sobre a
metodologia. Calculamos o tempo de chegada de um pedido subtraindo a
data de entrega da data de aprovação de uma ordem. Consideramos apenas
as ordens entregues, isto é, ordem com status igual a delivered. Foram
também removidas da conta as ordens com data de aprovação maior do que a
data de entrega.

Os pedidos levam em média 11 dias para chegar. Note a média é maior do
que a media, indicando a presença de outliers e que a distribuição é
assimétrica à direita. Isso significa que o tempo de entrega dos pedidos
se concentram mais entre 0 (mínimo) e 9 (mediana) dias e a existência de
pedidos que levam mais do que 15 (terceiro quartil) dias para chegar.

```python
def haversine(lat1, lon1, lat2, lon2):
    # Converter de graus para radianos
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    ## raio médio da terra
    R = 6371.0

    return R * c
```

```python
## geodesic( (-23.494428, -46.364439), (-20.519565, -47.391123)  )
distancias['distance_km'] = haversine(distancias['geolocation_lat'],
                                      distancias['geolocation_lng'],
                                      distancias['geolocation_lat.1'],
                                      distancias['geolocation_lng.1'])
```

## 3.2 Predição de atraso

Iremos considerar um pedido atrasado se a data de entrega é maior do que
a data estimada de entrega. Em outras palavras, uma pedido é atrasado se
*order_delivered_customer_date* > *order_estimated_delivery_date*.

```python
ordens = aquisicao_dados(processado_dir / 'olist_orders_dataset.csv',
                         mapeamento_dados)
```

```python
mascara = ordens['order_delivered_customer_date'] > ordens['order_estimated_delivery_date']
ordens[mascara]
```

```python
def verifica_atrasado(linha):
    if linha['order_delivered_customer_date'] > linha['order_estimated_delivery_date']:
        return 1

    return 0
```

```python
ordens['atrasado'] = ordens.apply(lambda linha: verifica_atrasado(linha), axis=1)
```

```python
mascara = ordens['order_delivered_customer_date'] > ordens['order_estimated_delivery_date']
ordens[mascara]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
ordens_transformado = ordens.drop(columns=['order_id', 'customer_id'])
encoder = LabelEncoder()
ordens_transformado['order_status'] = encoder.fit_transform(ordens_transformado['order_status'])
ordens_transformado
```

```python
colunas_data = ordens_transformado.select_dtypes(include='datetime').columns

for coluna in colunas_data:
    ordens_transformado[f'{coluna}_ano'] = ordens_transformado[coluna].dt.year
    ordens_transformado[f'{coluna}_mes'] = ordens_transformado[coluna].dt.month
    ordens_transformado.drop(columns=[coluna], inplace=True)
```

```python
ordens_transformado
```

```python
X = ordens_transformado.drop(columns=['atrasado'])
y = ordens_transformado['atrasado']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classificador_atrasado = RandomForestClassifier()
classificador_atrasado.fit(X_train, y_train)
```

```python
y_previsto = classificador_atrasado.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
print(classification_report(y_test, y_previsto))
```

```python
matriz_confusao = confusion_matrix(y_test, y_previsto)
sns.heatmap(matriz_confusao, annot=True)
```

Embora seja um modelo simples, ele obteve um bom desempenho em classificar
um pedido como atrasado (classe 1), porém tivemos uma revocação (recall)
baixo. Isso signfica que tivemos muitos pedidos classificados como
atrasados erroneamente (são classe 0 ao invés de 1). Se desejamos ter
